```python
import pandas as pd
import json
import numpy
```

```python
results_df = pd.read_csv("federalelections2016_bystate.csv")
```

```python
results_df.head()
```

```python
results_df = results_df.drop([0, 1, 2, 54, 55, 56, 57, 58, 59, 60], axis=0)
```

```python
results_df = results_df.drop('Unnamed: 7', 1)
```

```python
results_df
```

```python
results_df = results_df.rename(columns={'2016 PRESIDENTIAL ELECTORAL AND POPULAR VOTE': 'State', 'Unnamed: 1': 'EV_Rep', 'Unnamed: 2': 'EV_Dem','Unnamed: 3': 'PV_Rep','Unnamed: 4': 'PV_Dem','Unnamed: 5': 'PV_Other','Unnamed: 6': 'PV_Total',})
```

```python
results_df=results_df.reset_index(drop=True)
```

```python
results_df
```

```python
abb_df = pd.read_csv("state_abb_list.csv")
```

```python
abb_df
```

```python
combined_df = pd.merge(results_df, abb_df, how='left', left_on='State', right_on='Abb')
```

```python
combined_df
```

```python
combined_df = combined_df.drop('Abb', 1)
```

```python
combined_df = combined_df.rename(columns={'State_x': 'Abb', 'State_y': "State"})
```

```python
combined_df.loc[8, "State"] = "District of Columbia"
```

```python
combined_df
```

```python
combined_df.dtypes
```

```python
# Cast objects as strings
combined_df['PV_Rep'] = combined_df['PV_Rep'].str.replace(",", "")
combined_df['PV_Dem'] = combined_df['PV_Dem'].str.replace(",", "")
combined_df['PV_Other'] = combined_df['PV_Other'].str.replace(",", "")
combined_df['PV_Total'] = combined_df['PV_Total'].str.replace(",", "")
```

```python
combined_df.dtypes
```

```python
combined_df['PV_Rep'] = combined_df['PV_Rep'].astype(int)
combined_df['PV_Dem'] = combined_df['PV_Dem'].astype(int)
combined_df['PV_Other'] = combined_df['PV_Other'].astype(int)
combined_df['PV_Total'] = combined_df['PV_Total'].astype(int)
```

```python
combined_df['Per_Rep'] = round(combined_df['PV_Rep'] / combined_df['PV_Total'] * 100, 2)
combined_df['Per_Dem'] = round(combined_df['PV_Dem'] / combined_df['PV_Total'] * 100, 2)
combined_df['Per_Other'] = round(combined_df['PV_Other'] / combined_df['PV_Total'] * 100, 2)
```

```python
combined_df.to_csv ("../static/js/cleaned_federal_elections.csv")
```

```python
result = combined_df.to_json(orient="table")
parsed = json.loads(result)
json.dumps(parsed, indent=4)
```

```python
import pandas as pd
import json
```

```python
covid_df = pd.read_csv("covid_data.csv")
```

```python
unemployement_df = pd.read_csv("unemployment.csv")
```

```python
covid_df.head()
```

```python
unemployement_df.head()
```

```python
combined_df =  pd.merge(unemployement_df, covid_df, how='left', left_on=['state', 'year', 'month'], right_on=['state', 'year', 'month'])
```

```python
combined_df
```

```python
combined_df['total_cases'] = combined_df['total_cases'].fillna(0)
combined_df['total_deaths'] = combined_df['total_deaths'].fillna(0)
```
